```python
from pdfminer.high_level import extract_text
from docx import Document

def parse_pdf(file_path):
    return extract_text(file_path)

def parse_docx(file_path):
    doc = Document(file_path)
    return "\n".join([p.text for p in doc.paragraphs])
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
def generate_embedding(text):
    return model.encode(text, show_progress_bar=False)
```

```python
from fastapi import FastAPI, UploadFile, HTTPException
import asyncio

app = FastAPI()
chromadb_client = get_chromadb_client()
```

```python
!pip install python-multipart
```

```python
from fastapi import BackgroundTasks

@app.post("/ingest/")
async def ingest_document(file: UploadFile, background_tasks: BackgroundTasks):
    # Parse document
    content = await file.read()
    # Save content asynchronously
    background_tasks.add_task(save_document, content)
    return {"status": "Document ingestion started"}
```

```python
@app.get("/query/")
async def query_documents(query: str):
    embedding = generate_embedding(query)
    results = chromadb_client.query(embedding)  # Search ChromaDB
    return {"results": results}
```

```python
async def save_document(content):
    # Generate embeddings
    embedding = generate_embedding(content)
    # Store in ChromaDB
    chromadb_client.insert(embedding)
```
